```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
neon = {
    'Длины волн линий спектра неона, нм': [640.2, 614.3, 594.5, 585.2, 576, 540],
    'Градусные деления барабана': [],
}
mercury = {
    'Длины волн линий спектра ртути, нм': [576.9, 573.1, 546.1, 491.6, 435.6, 404.6],
    'Градусные деления барабана': [],
}
```

```python
neon['Градусные деления барабана'] = [int(input("Введите град. дел. бар. при длине волны неона {} нм => ".format(neon['Длины волн линий спектра неона, нм'][i]))) for i in range(6)]
```

```python
mercury['Градусные деления барабана'] = [int(input("Введите град. дел. бар. при длине волны неона {} нм => ".format(mercury['Длины волн линий спектра ртути, нм'][i]))) for i in range(6)]
```

```python
mercuryX = mercury['Градусные деления барабана']
mercuryY = mercury['Длины волн линий спектра ртути, нм']
neonX = neon['Градусные деления барабана']
neonY = neon['Длины волн линий спектра неона, нм']
```

```python
modelMercuryX = np.array(mercuryX).reshape((-1, 1))
modelMercuryY = [int(mercuryY[i]) for i in range(6)]

modelMercury = LinearRegression(fit_intercept = True)

modelMercury.fit(modelMercuryX, modelMercuryY)

approxMercuryX = np.linspace(850, 2050)
approxMercuryY = modelMercury.predict(approxMercuryX[:, np.newaxis])
```

```python
modelNeonX = np.array(neonX).reshape((-1, 1))
modelNeonY = [int(neonY[i]) for i in range(6)]

modelNeon = LinearRegression(fit_intercept = True)

modelNeon.fit(modelNeonX, modelNeonY)

approxNeonX = np.linspace(2051, 2450)
approxNeonY = modelNeon.predict(approxNeonX[:, np.newaxis])
```

```python
plt.figure(figsize=(20, 10))

plt.title("Градуированный график")
plt.xlabel("Градусные деления барабана")
plt.ylabel("Длина волны, нм")
plt.grid()

plt.scatter(mercuryX, mercuryY, label='Ртуть')
plt.plot(approxMercuryX, approxMercuryY, label = "Апроксимация ртути", color='red')

plt.scatter(neonX, neonY, label='Неон')
plt.plot(approxNeonX, approxNeonY, label = "Апроксимация неона", color = 'green')

plt.legend()
```
